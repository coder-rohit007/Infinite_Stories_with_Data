# What is culmen?

The upper margin of the beak or bill is referred to as the culmen and the measurement is taken using calipers with one jaw at the tip of the upper mandible and the other at base of the skull or the first feathers depending on the standard chosen.

# Which species of penguin have highest mean culmen length? ---> Chinstrap

```python
df = data.groupby('species')['culmen_length_mm'].mean().reset_index()

plt.figure(figsize=(10,7))
fig = sns.barplot(data=df, x='culmen_length_mm', y='species')
plt.show()
```

# Culmen Length distribution

```python
sns.violinplot(data=data, x="species", y="culmen_length_mm", size=8)
plt.show()
```

# culmen_length vs culmen_depth

```python
sns.scatterplot(data=data, x='culmen_length_mm', y='culmen_depth_mm', hue='species')
plt.show()
```

# Which island have highest count of penguins? ---> Biscoe

```python
data['island'].value_counts().plot(kind='barh')
plt.show()
```

# Pairplot

```python
sns.pairplot(data=data[['species','culmen_length_mm','culmen_depth_mm','flipper_length_mm', 'body_mass_g']], hue="species", size=3, diag_kind="hist")
plt.show()
```
